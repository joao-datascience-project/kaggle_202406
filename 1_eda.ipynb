# Univariate Evaluation of each Feature

```python
train_data.columns
```

```python
train_data['Marital status'].isna().sum()
```

```python
train_data['Marital status'].value_counts()
```

```python
ax = train_data['Marital status'].value_counts().sort_index().plot(kind='bar',rot=0)

# iterate over the bars and annotate each one
for p in ax.containers:
    for q in p:
        ax.annotate(str(round(q.get_height(), 2)),
                    (q.get_x() + q.get_width() / 2., q.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.title(f"relative frequencies for each label")
```

```python
ax = train_data['Marital status'].value_counts(normalize=True).sort_index().plot(kind='bar',rot=0)

# iterate over the bars and annotate each one
for p in ax.containers:
    for q in p:
        ax.annotate(str(round(q.get_height(), 4)),
                    (q.get_x() + q.get_width() / 4., q.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.title(f"Proportions for each label")
```

```python
train_data.groupby('Marital status')['Target'].value_counts(normalize=True).unstack().plot(kind='bar', stacked=True, rot=True, title='Target proportion for each label')
```
